```python
def get_caredetails(auto):
    # Check if the car name (index 8) matches
    return auto[8].strip() == 'chevrolet chevelle malibu'

def run():
    with beam.Pipeline() as pipeline:
        car_records = (
            pipeline
            | "Read from Text File" >> beam.io.ReadFromText('auto-mpg.txt', skip_header_lines=1)
            | "Split Record by Comma" >> beam.Map(lambda record: record.split(','))
            | "Filter for chevrolet chevelle malibu" >> beam.Filter(get_caredetails)
        )

        # Display matching records in the console
        car_records | "Print Filtered Records" >> beam.Map(print)

        # Write matching records to output file
        (
            car_records
            | "Convert List to CSV String" >> beam.Map(lambda record: ','.join(record))
            | "Write to Text File" >> beam.io.WriteToText('chevrolet12', num_shards=1)
        )

if __name__ == "__main__":
    run()
```
